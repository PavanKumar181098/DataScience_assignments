```python
data_new=data.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
```

```python
data_new
```

```python
data_new.index=data['User.ID'].unique()
```

```python
data_new
```

```python
#cosine similarity
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
data_new.fillna(0,inplace=True)
```

```python
user_sim=1-pairwise_distances(data_new.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index=data['User.ID'].unique()
user_sim_df.columns=data['User.ID'].unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```
