# Hierarchical Clustering

```python
#import hierarchichal clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime=pd.read_csv("/content/crime_data.csv")
```

```python
#normalization
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
#normalized data frame
df_norm=norm_func(crime.iloc[:,1:])
```

```python
#create dendrogram
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
#there are two clusters as per the dendrogram
```

```python
#create clusters after examining the dendrogram
hc=AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
```

```python
#save clusters for chart
y_hc=hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc, columns=['Clusters'])
```

```python
Clusters
```

```python
df_norm['h_clusterid']=Clusters
```

```python
df_norm
```

```python
crime['h_clusterid']=Clusters
```

```python
crime
```

```python
#hierarchical clustering done; no.of clusters=2
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
crime_k=pd.read_csv("/content/crime_data.csv")
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_crime_df=scaler.fit_transform(crime_k.iloc[:,1:])
```

```python
#finding the optimum no. of cluster or kvalue
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(scaled_crime_df)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build cluster algorithm
from sklearn.cluster import KMeans
clusters_new=KMeans(4, random_state=2)#choosing no.of clusters as 5 from the above elbow graph
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
#assign clustes to the data set
crime_k['clusterid_new']=clusters_new.labels_
```

```python
crime_k #dataset arranged along with clusters
```

```python
crime_k.groupby('clusterid_new').agg(['mean']).reset_index
```

# DBSCAN

```python
crime_db=pd.read_csv("/content/crime_data.csv")
```

```python
from sklearn.cluster import DBSCAN
crime_db.head()
```

```python
crime_db.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
array=crime_db.values
```

```python
array
```

```python
stscaler=StandardScaler().fit(array)
X=stscaler.transform(array)
```

```python
X
```

```python
dbscan=DBSCAN(eps=1.0,min_samples=4)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
cl
```

```python
pd.concat([crime_db,cl],axis=1)
```

```python
crime_db[['clusters']]=cl
```

```python
crime_db
```

```python
crime_db.dtypes
```

```python
crime_db_new=crime_db[crime_db['clusters']!=-1] #filtering out all the outliers from the dbscan to apply the kmeans clustering
```

```python
crime_db_new
```

```python
crime_db_new=crime_db_new.drop(['clusters'],axis=1)
```

```python
crime_db_new
```

```python
scaler1=StandardScaler()
scaled_crime_df1=scaler1.fit_transform(crime_db_new.iloc[:,:])
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(scaled_crime_df1)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new1=KMeans(4, random_state=42)#choosing no.of clusters as 5 from the above elbow graph
clusters_new1.fit(scaled_crime_df1)
```
